# Código para obtener las tétradas de homeólogos del genoma de _Fragaria x ananassa_

```python
#Download LAST from repository and installation
!wget https://gitlab.com/mcfrith/last/-/archive/1256/last-1256.zip
!unzip last-1256.zip
!cd last-1256; make install
```

```python
#Download the CDS of each subgenome and decompress them
!wget https://github.com/agm-bioinfo/Fragaria-x-ananassa-RNA-seq/raw/main/Homeologs/A.zip
!wget https://github.com/agm-bioinfo/Fragaria-x-ananassa-RNA-seq/raw/main/Homeologs/B.zip
!wget https://github.com/agm-bioinfo/Fragaria-x-ananassa-RNA-seq/raw/main/Homeologs/C.zip
!wget https://github.com/agm-bioinfo/Fragaria-x-ananassa-RNA-seq/raw/main/Homeologs/D.zip

!unzip A.zip
!unzip B.zip
!unzip C.zip
!unzip D.zip
```

```python
#Create databases from each subgenome
!lastdb -uNEAR -R01 -P0 A_db A.fa
!lastdb -uNEAR -R01 -P0 B_db B.fa
!lastdb -uNEAR -R01 -P0 C_db C.fa
!lastdb -uNEAR -R01 -P0 D_db D.fa
```

```python
#Align each subgenome to every other subgenome
#Not all alignments are necesary but can be useful if you have gff files and sequences coordinates and need them
#cause alignments A_vs_B and B_vs_A will report diferent coordinates based on the first database
!lastal A_db -P0 -D10000000000 B.fa | last-map-probs -m 0.49 > A.B.maf
!lastal A_db -P0 -D10000000000 C.fa | last-map-probs -m 0.49 > A.C.maf
!lastal A_db -P0 -D10000000000 D.fa | last-map-probs -m 0.49 > A.D.maf

!lastal B_db -P0 -D10000000000 A.fa | last-map-probs -m 0.49 > B.A.maf
!lastal B_db -P0 -D10000000000 C.fa | last-map-probs -m 0.49 > B.C.maf
!lastal B_db -P0 -D10000000000 D.fa | last-map-probs -m 0.49 > B.D.maf

!lastal C_db -P0 -D10000000000 B.fa | last-map-probs -m 0.49 > C.B.maf
!lastal C_db -P0 -D10000000000 A.fa | last-map-probs -m 0.49 > C.A.maf
!lastal C_db -P0 -D10000000000 D.fa | last-map-probs -m 0.49 > C.D.maf

!lastal D_db -P0 -D10000000000 B.fa | last-map-probs -m 0.49 > D.B.maf
!lastal D_db -P0 -D10000000000 C.fa | last-map-probs -m 0.49 > D.C.maf
!lastal D_db -P0 -D10000000000 D.fa | last-map-probs -m 0.49 > D.A.maf
```

```python
#Download scripts from https://github.com/tony-kuo/eagle/tree/master/scripts to get best reciprocal hits between alignments and generate groups of 4 homeologs from each subgenome
!wget https://raw.githubusercontent.com/tony-kuo/eagle/master/scripts/homeolog_genotypes.py
!wget https://raw.githubusercontent.com/tony-kuo/eagle/master/scripts/quad_homeolog.pl
```

```python
#Generate best hits between the alignments
!python homeolog_genotypes.py -o A.vs.B A.B.maf B.A.maf
!python homeolog_genotypes.py -o A.vs.C A.C.maf C.A.maf
!python homeolog_genotypes.py -o A.vs.D A.D.maf D.A.maf

!python homeolog_genotypes.py -o B.vs.A B.A.maf A.B.maf
!python homeolog_genotypes.py -o B.vs.C B.C.maf C.B.maf
!python homeolog_genotypes.py -o B.vs.D B.D.maf D.B.maf

!python homeolog_genotypes.py -o C.vs.B C.B.maf B.C.maf
!python homeolog_genotypes.py -o C.vs.A C.A.maf A.C.maf
!python homeolog_genotypes.py -o C.vs.D C.D.maf D.C.maf

!python homeolog_genotypes.py -o D.vs.B D.B.maf B.D.maf
!python homeolog_genotypes.py -o D.vs.C D.C.maf C.D.maf
!python homeolog_genotypes.py -o D.vs.A D.A.maf A.D.maf
```

```python
!perl quad_homeolog.pl A.vs.B.reciprocal_best A.vs.C.reciprocal_best A.vs.D.reciprocal_best B.vs.C.reciprocal_best B.vs.D.reciprocal_best C.vs.D.reciprocal_best > homeolog.ABCD.list
```
